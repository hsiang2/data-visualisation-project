```python
for loc in range(19):
    df = dfs_by_location[loc+1].copy()
    
    df['time'] = pd.to_datetime(df['time'])

    df['time_bin'] = df['time'].dt.floor('1h')

    report_counts = df.groupby('time_bin').size().reset_index(name='count')

    full_range = pd.date_range(start=report_counts['time_bin'].min(),
                               end=report_counts['time_bin'].max(),
                               freq='1h')

    report_counts_full = pd.DataFrame({'time_bin': full_range}).merge(
        report_counts, on='time_bin', how='left').fillna(0)

    plt.figure(figsize=(12, 5))
    sns.lineplot(x='time_bin', y='count', data=report_counts_full, marker='o')
    plt.xticks(rotation=45)
    plt.title(f'Report Counts per 1 Hours (Community {loc+1})')
    plt.xlabel('Time')
    plt.ylabel('Number of Reports')

    for x, y in zip(report_counts_full['time_bin'], report_counts_full['count']):
        plt.text(x, y + 0.3, str(int(y)), ha='center', va='bottom', fontsize=9)

    plt.tight_layout()
    plt.grid(True, axis='y')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

colors = sns.color_palette("tab20", n_colors=19)
markers = ['o', 's', 'D', '^', 'v', '>', '<', 'X', 'P', '*']

plt.figure(figsize=(16, 7))

for i in range(19):
    loc = i + 1
    df = dfs_by_location[loc].copy()
    df['time'] = pd.to_datetime(df['time'])
    df['time_bin'] = df['time'].dt.floor('1h')
    report_counts = df.groupby('time_bin').size().reset_index(name='count')
    
    full_range = pd.date_range(start=report_counts['time_bin'].min(),
                               end=report_counts['time_bin'].max(),
                               freq='1h')
    
    report_counts_full = pd.DataFrame({'time_bin': full_range}).merge(
        report_counts, on='time_bin', how='left').fillna(0)

    color = colors[i % len(colors)]
    marker = markers[i % 10]  

    sns.lineplot(
        x='time_bin',
        y='count',
        data=report_counts_full,
        label=f'Community {loc}',
        color=color,
        marker=marker
    )

plt.xticks(rotation=45)
plt.title('Report Counts per 1 Hours for All Communities')
plt.xlabel('Time')
plt.ylabel('Number of Reports')
plt.legend(title='Community', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.grid(True, axis='y')
plt.savefig("time-report.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json

if not dfs_by_location:
    print("Error: dfs_by_location dictionary is empty. Please load your data first.")
    # raise ValueError("Data not loaded into dfs_by_location")
else:
    print("Processing data for Matplotlib plot and JSON export (with zero-filling)...")

    plotly_data_traces = [] 
    matplotlib_data_to_plot = {} 

    markers_plotly = ['circle', 'square', 'diamond', 'cross', 'x', 'triangle-up', 'triangle-down', 'pentagon', 'hexagon', 'star']
    markers_mpl = ['o', 's', 'D', '^', 'v', '>', '<', 'X', 'P', '*'] 

    all_original_time_bins = set()
    valid_location_data = {} 

    for i in range(19):
        loc = i + 1
        if loc not in dfs_by_location:
            continue
        df = dfs_by_location[loc].copy()
        if 'time' not in df.columns:
             continue
        try:
            df['time'] = pd.to_datetime(df['time'])
            df['time_bin'] = df['time'].dt.floor('1h')
            report_counts = df.groupby('time_bin').size().reset_index(name='count')
            if not report_counts.empty:
                all_original_time_bins.update(report_counts['time_bin'])
                valid_location_data[loc] = report_counts 
        except Exception as e:
            print(f"Warning: Error processing time data for location {loc}: {e}")
            continue

    if not all_original_time_bins:
        print("Error: No valid time data found across all locations.")
    else:
        global_min_time = min(all_original_time_bins)
        global_max_time = max(all_original_time_bins)
        full_range_global = pd.date_range(start=global_min_time, end=global_max_time, freq='1h')
        full_range_df = pd.DataFrame({'time_bin': full_range_global})

        for i in range(19):
            loc = i + 1
            if loc in valid_location_data:
                report_counts = valid_location_data[loc]

                report_counts_full = full_range_df.merge(
                    report_counts, on='time_bin', how='left').fillna(0)

                matplotlib_data_to_plot[loc] = report_counts_full

                trace_data = {
                    'x': report_counts_full['time_bin'].dt.strftime('%Y-%m-%d %H:%M:%S').tolist(),
                    'y': report_counts_full['count'].tolist(),
                    'mode': 'lines+markers',
                    'name': f'Community {loc}',
                    'type': 'scatter',
                    'marker': {'symbol': markers_plotly[i % len(markers_plotly)]}
                }
                plotly_data_traces.append(trace_data)

        output_json_path = "community_hourly_reports.json"
        final_json_data = {"traces": plotly_data_traces}
        try:
            with open(output_json_path, 'w') as f:
                json.dump(final_json_data, f, indent=4)
            print(f"Successfully exported data (with zero-filling) to {output_json_path}")
        except Exception as e:
            print(f"Error exporting data to JSON: {e}")


        print("Generating Matplotlib plot (using zero-filled data)...")
        plt.figure(figsize=(16, 7))
        colors = sns.color_palette("tab20", n_colors=19)

        plotted_count_mpl = 0
        for i in range(19):
             loc = i + 1
             if loc in matplotlib_data_to_plot:
                report_counts_full_mpl = matplotlib_data_to_plot[loc]
                color = colors[i % len(colors)]
                marker = markers_mpl[i % 10] 

                sns.lineplot(
                    x='time_bin',
                    y='count',
                    data=report_counts_full_mpl, 
                    label=f'Community {loc}',
                    color=color,
                    marker=marker
                )
                plotted_count_mpl += 1

        print(f"Plotted {plotted_count_mpl} communities in Matplotlib.")

        plt.xticks(rotation=45)
        plt.title('Report Counts per 1 Hour for All Communities (Zero-Filled)')
        plt.xlabel('Time')
        plt.ylabel('Number of Reports')
        plt.legend(title='Community', bbox_to_anchor=(1.02, 1), loc='upper left', ncol=2)
        plt.tight_layout(rect=[0, 0, 0.9, 1])
        plt.grid(True, axis='y')
        plt.show()
```

```python
df = dfs_by_location[8].copy()
df['time'] = pd.to_datetime(df['time'])

df = df.sort_values('time').set_index('time')

rolling_mean = df['sewer_and_water'].rolling('6h').mean()
rolling_std = df['sewer_and_water'].rolling('6h').std()

rolling_stats = pd.DataFrame({
    'mean': rolling_mean,
    'std': rolling_std
})

plt.figure(figsize=(16, 5))
sns.lineplot(data=rolling_stats)
plt.title('6-Hour Rolling Mean and Std of sewer_and_water (Community 1)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, axis='y')
plt.savefig("rolling_mean_std.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = dfs_by_location[4].copy()

df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time').sort_index()

target_columns = ['sewer_and_water']

corr_results = {}
for col in target_columns:
    corr_series = df[col].rolling('1h', min_periods=20).corr(df['shake_intensity'])
    corr_results[col] = corr_series

corr_df = pd.DataFrame(corr_results)

plt.figure(figsize=(12, 6))

for col in target_columns:
    plt.plot(corr_df.index, corr_df[col], label=col)

plt.xlabel('Time')
plt.ylabel('Correlation Coefficient')
plt.title('Rolling 12-Hour Correlation with Shake Intensity (min 50 points)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"data num per hour: {df.resample('1H').count().mean().mean():.1f}")
```
